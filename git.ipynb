```python
import requests
import pandas as pd

headers = {"Authorization": "token your_personal_access_token"}

# Function to fetch repositories for a given topic
def fetch_github_repositories(topic):
    url = f"https://api.github.com/search/repositories?q=topic:{topic}&sort=stars&order=desc"
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.json()['items']
    else:
        print(f"Failed to fetch data for {topic}: {response.status_code}")
        return []

# List of trending topics
topics = ['machine learning', 'data visualization', 'deep learning', 'natural language processing']

# Extracting data for each topic
repositories_data = []
for topic in topics:
    repositories_data.extend(fetch_github_repositories(topic))

# Converting to a Pandas DataFrame
df = pd.DataFrame(repositories_data)

# Select the fields you want to store
df_cleaned = df[['name', 'owner', 'description', 'html_url', 'language', 'created_at', 'updated_at', 'stargazers_count', 'forks_count', 'open_issues_count', 'license']]

# Clean the data
df_cleaned['owner'] = df_cleaned['owner'].apply(lambda x: x['login'])
df_cleaned['license'] = df_cleaned['license'].apply(lambda x: x['name'] if x else 'No License')
df_cleaned['created_at'] = pd.to_datetime(df_cleaned['created_at'])
df_cleaned['updated_at'] = pd.to_datetime(df_cleaned['updated_at'])

# Renaming columns for better readability
df_cleaned.columns = ['Repository_Name', 'Owner', 'Description', 'URL', 'Programming_Language', 'Creation_Date', 'Last_Updated_Date', 'Number_of_Stars', 'Number_of_Forks', 'Number_of_Open_Issues', 'License_Type']

# Save to CSV for easy access later
df_cleaned.to_csv('repositories_data.csv', index=False)

# Print the top 5 records
print(df_cleaned.head())
```
