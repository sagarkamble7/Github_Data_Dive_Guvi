```python
import mysql.connector
import pandas as pd

# Load the data from CSV
df_cleaned = pd.read_csv(r'C:\Users\sagar\Desktop\GIT_REPO_PROJECT\repo_data.csv')

# Drop the Unnamed column (assuming it's the first column)
df_cleaned.drop(columns=['Unnamed: 0'], inplace=True)

# Connect to MySQL database
db = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Sagar72427',
    port=3306,
    auth_plugin='mysql_native_password',
    database="github_data"
)

cursor = db.cursor()

# SQL Insert query
insert_query = """
INSERT INTO repositories (
    Repository_Name,
    Owner,
    Description,
    URL,
    Programming_Language,
    Creation_Date,
    Last_Updated_Date,
    Number_of_Stars,
    Number_of_Forks,
    Number_of_Open_Issues,
    License_Type
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Prepare data for bulk insert
data_to_insert = []

for index, row in df_cleaned.iterrows():
    if row.isnull().any():
        print(f"Row {index} contains NaN values: {row}")  # Identify problematic rows
        continue  # Skip this row or handle it as needed
    row_data = tuple(row)
    print("Preparing Row for Insert:", row_data)  # Debugging statement
    data_to_insert.append(row_data)

# Execute the insert
cursor.executemany(insert_query, data_to_insert)

# Commit the changes
db.commit()

# Close the connection
cursor.close()
db.close()

print("Data inserted successfully!")
```
